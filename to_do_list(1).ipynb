```python
import ipywidgets as widgets
from IPython.display import display, clear_output
```

```python
tasks = []

FILENAME = "task.txt"
```

```python
# Display tasks simply
def list_tasks():
    if not tasks:
        print("📭 Henüz görev yok.")
    else:
        print("📋 Görev Listesi:")
        for i, task in enumerate(tasks, 1):
            print(f"{i}. {task}")
```

```python
def add_task_interface():
    clear_output()
    task_input = widgets.Text(placeholder="Yeni görev girin")
    add_button = widgets.Button(description="Görevi Ekle", button_style="success")

    def add_task(b):
        if task_input.value.strip():
            tasks.append(task_input.value.strip())
            print("✅ Görev eklendi.\n")
            task_input.value = ""
        else:
            print("❌ Boş görev girilemez.")
        list_tasks()
        show_menu()

    add_button.on_click(add_task)
    display(task_input, add_button)
```

```python
def delete_task_interface():
    clear_output()

    if not tasks:
        print("📭 Silinecek görev yok.")
        show_menu()
        return

    options = [f"{i+1}. {tasks[i]}" for i in range(len(tasks))]
    dropdown = widgets.Dropdown(options=options, description="Sil:")
    delete_button = widgets.Button(description="Sil", button_style="danger")

    def delete_task(b):
        clear_output()
        selected = dropdown.index
        deleted = tasks.pop(selected)
        print(f"🗑️ Silindi: {deleted}\n")

        if not tasks:
            print("📭 Tüm görevler silindi.")
        else:
            list_tasks()

        show_menu()

    delete_button.on_click(delete_task)
    display(dropdown, delete_button)
```

```python
def edit_task_interface():
    clear_output()

    if not tasks:
        print("📭 Düzenlenecek görev yok.")
        show_menu()
        return

    options = [f"{i+1}. {tasks[i]}" for i in range(len(tasks))]
    dropdown = widgets.Dropdown(options=options, description="Seç:")
    new_text = widgets.Text(placeholder="Yeni görev metni")
    edit_button = widgets.Button(description="Güncelle", button_style="info")

    def update_task(b):
        clear_output()
        if new_text.value.strip():
            tasks[dropdown.index] = new_text.value.strip()
            print("✏️ Görev güncellendi.\n")
        else:
            print("❌ Boş görev girilemez.\n")
        list_tasks()
        show_menu()

    edit_button.on_click(update_task)
    display(dropdown, new_text, edit_button)
```

```python
def show_menu():
    print("📌 Ana Menü")
    add_btn = widgets.Button(description="Yeni Görev Ekle", button_style="success")
    delete_btn = widgets.Button(description="Görev Sil", button_style="danger")
    edit_btn = widgets.Button(description="Görev Düzenle", button_style="info")
    list_btn = widgets.Button(description="Görevleri Listele", button_style="primary")

    def add(b):
      add_task_interface()
    def delete(b):
      delete_task_interface()
    def edit(b):
      edit_task_interface()
    def list_all(b):
        clear_output()
        list_tasks()
        show_menu()

    add_btn.on_click(add)
    delete_btn.on_click(delete)
    edit_btn.on_click(edit)
    list_btn.on_click(list_all)

    display(add_btn, delete_btn, edit_btn, list_btn)

```

```python
show_menu()
```
