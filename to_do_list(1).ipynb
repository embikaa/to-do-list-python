```python
import ipywidgets as widgets
from IPython.display import display, clear_output
```

```python
tasks = []

FILENAME = "task.txt"
```

```python
# Display tasks simply
def list_tasks():
    if not tasks:
        print("üì≠ Hen√ºz g√∂rev yok.")
    else:
        print("üìã G√∂rev Listesi:")
        for i, task in enumerate(tasks, 1):
            print(f"{i}. {task}")
```

```python
def add_task_interface():
    clear_output()
    task_input = widgets.Text(placeholder="Yeni g√∂rev girin")
    add_button = widgets.Button(description="G√∂revi Ekle", button_style="success")

    def add_task(b):
        if task_input.value.strip():
            tasks.append(task_input.value.strip())
            print("‚úÖ G√∂rev eklendi.\n")
            task_input.value = ""
        else:
            print("‚ùå Bo≈ü g√∂rev girilemez.")
        list_tasks()
        show_menu()

    add_button.on_click(add_task)
    display(task_input, add_button)
```

```python
def delete_task_interface():
    clear_output()

    if not tasks:
        print("üì≠ Silinecek g√∂rev yok.")
        show_menu()
        return

    options = [f"{i+1}. {tasks[i]}" for i in range(len(tasks))]
    dropdown = widgets.Dropdown(options=options, description="Sil:")
    delete_button = widgets.Button(description="Sil", button_style="danger")

    def delete_task(b):
        clear_output()
        selected = dropdown.index
        deleted = tasks.pop(selected)
        print(f"üóëÔ∏è Silindi: {deleted}\n")

        if not tasks:
            print("üì≠ T√ºm g√∂revler silindi.")
        else:
            list_tasks()

        show_menu()

    delete_button.on_click(delete_task)
    display(dropdown, delete_button)
```

```python
def edit_task_interface():
    clear_output()

    if not tasks:
        print("üì≠ D√ºzenlenecek g√∂rev yok.")
        show_menu()
        return

    options = [f"{i+1}. {tasks[i]}" for i in range(len(tasks))]
    dropdown = widgets.Dropdown(options=options, description="Se√ß:")
    new_text = widgets.Text(placeholder="Yeni g√∂rev metni")
    edit_button = widgets.Button(description="G√ºncelle", button_style="info")

    def update_task(b):
        clear_output()
        if new_text.value.strip():
            tasks[dropdown.index] = new_text.value.strip()
            print("‚úèÔ∏è G√∂rev g√ºncellendi.\n")
        else:
            print("‚ùå Bo≈ü g√∂rev girilemez.\n")
        list_tasks()
        show_menu()

    edit_button.on_click(update_task)
    display(dropdown, new_text, edit_button)
```

```python
def show_menu():
    print("üìå Ana Men√º")
    add_btn = widgets.Button(description="Yeni G√∂rev Ekle", button_style="success")
    delete_btn = widgets.Button(description="G√∂rev Sil", button_style="danger")
    edit_btn = widgets.Button(description="G√∂rev D√ºzenle", button_style="info")
    list_btn = widgets.Button(description="G√∂revleri Listele", button_style="primary")

    def add(b):
      add_task_interface()
    def delete(b):
      delete_task_interface()
    def edit(b):
      edit_task_interface()
    def list_all(b):
        clear_output()
        list_tasks()
        show_menu()

    add_btn.on_click(add)
    delete_btn.on_click(delete)
    edit_btn.on_click(edit)
    list_btn.on_click(list_all)

    display(add_btn, delete_btn, edit_btn, list_btn)

```

```python
show_menu()
```
